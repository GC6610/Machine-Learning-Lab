```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df =  pd.read_excel("Concrete_Data.xls")
df.head() 
```

```python
x = df.values[:, [0,7]]  
```

```python
def Scaling(x):
    mu = np.mean(x, axis = 0)  
    sigma = np.std(x, axis= 0, ddof = 1)  #standard deviation
    x_norm = (x - mu)/sigma
    return x_norm
#scaling down the values , i tried this since I was getting overflow error 
x= Scaling(x)
x
```

```python
#columns to take under consideration
x1=pd.Series(x[:,0])
x2=pd.Series(x[:,1])
y=df['Concrete compressive strength(MPa, megapascals) ']
```

```python
def cost(x1,x2,y,b,m1,m2):
    rows= len(x1)
    costs=0
    for i in range(rows):
        y_pred=(m1*x1[i]+m2*x2[i]+b)
        costs+=(abs(y_pred-y[i]))
#         cost+=((1/2*rows)*((y_pred-y[i])**2))
    return costs/rows
# cost(x1,x2,y,b,m1,m2)
```

```python
def step_gradient(x1,x2,y,b,m1,m2,alpha):
    m1_der=m1
    m2_der=m2
    b_der=b
    rows=len(x1)
    for i in range(rows):
        b_der+=(-2/rows)*(y[i]-(m1*x1[i]+m2*x2[i]+b))
        m1_der+=(-2/rows)*(x1[i])*(y[i]-(m1*x1[i]+b))
        m2_der+=(-2/rows)*(x2[i])*(y[i]-(m2*x2[i]+b))
    new_b=b-b_der*alpha
    new_m1=m1-m1_der*alpha
    new_m2=m2-m2_der*alpha
    return new_b,new_m1,new_m2
    
```

```python
def gradient_descent(x1,x2,y,alpha,steps):
    b=0
    m1=0
    m2=0
    coeff=[]
    costs=[]
    for i in range(steps):
        b,m1,m2 = step_gradient(x1,x2,y,b,m1,m2,alpha)
        coeff.append([b,m1,m2])
        get_cost=cost(x1,x2,y,b,m1,m2)
        costs.append(get_cost)
    min_cost_idx=costs.index(min(costs))
    opt_coeff=coeff[min_cost_idx]
    return costs,coeff,opt_coeff
        
    
```

```python
cost,coeff,opt_coeff=gradient_descent(x1,x2,y,0.01,100)
for i in cost:
    print(i)
```
