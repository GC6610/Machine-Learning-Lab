```python
#taking two columns sepal_length and petal_length as input cols
x1=pd.Series(x[:,0])
x2=pd.Series(x[:,2])
y=df['species']
```

```python
# Cost Function
def cost_cal(x1,x2,y,b,m1,m2):
    rows= len(x1)
    costs=0
    for i in range(rows):
        y_pred=(m1*x1[i]+m2*x2[i]+b)
        costs+=(1/(2*rows))*((y_pred-y[i])**2)
    return costs
# cost(x1,x2,y,b,m1,m2)
```

```python
def step_gradient(x1,x2,y,b,m1,m2,alpha):
    m1_der=m1
    m2_der=m2
    b_der=b
    rows=len(x1)
    for i in range(rows):
        b_der+=(-2/rows)*(y[i]-(m1*x1[i]+m2*x2[i]+b))
        m1_der+=(-2/rows)*(x1[i])*(y[i]-(m1*x1[i]+b))
        m2_der+=(-2/rows)*(x2[i])*(y[i]-(m2*x2[i]+b))
    new_b=b-b_der*alpha
    new_m1=m1-m1_der*alpha
    new_m2=m2-m2_der*alpha
    return new_b,new_m1,new_m2
```

```python
def gradient_descent(x1,x2,y,alpha,steps):
    b=0
    m1=0
    m2=0
    coeff=[]
    costs=[]
    for i in range(steps):
        b,m1,m2 = step_gradient(x1,x2,y,b,m1,m2,alpha)
        coeff.append([b,m1,m2])
        get_cost=cost_cal(x1,x2,y,b,m1,m2)
        costs.append(get_cost)
    min_cost_idx=costs.index(min(costs))
    opt_coeff=coeff[min_cost_idx]
    return costs,coeff,opt_coeff
        
```

```python
def predict(m1,m2,b,sep_len,pet_len):
    return round(m1*sep_len+m2*pet_len+b) 
    
        
    
```

```python
cost,coeff,opt_coeff=gradient_descent(x1,x2,y,0.01,500)
for i in cost:
    print(i)
```

```python
b,m1,m2=opt_coeff[0],opt_coeff[1],opt_coeff[2]
correct=0
for i in range(len(x1)):
    #predicted values
#     print(y[i],predict(m1,m2,b,x1[i],x2[i]))
    if y[i]==predict(m1,m2,b,x1[i],x2[i]):
        correct+=1
```

```python
print("Accuracy : ",correct*100/len(x1))
```
