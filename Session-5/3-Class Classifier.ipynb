```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import math
```

```python
def sigmoid(z):
    return 1.0/(1 + np.exp(-z))
```

```python
def loss(y, y_sigpred):
    loss = -np.mean(y*(np.log(y_sigpred)) - (1-y)*np.log(1-y_sigpred))
    return loss
```

```python

def gradients(X, y, y_sigpred):
    
    # X --> Input.
    # y --> true/target value.
    # y_sigpred --> hypothesis/predictions.
    # w --> weights (parameter).
    # b --> bias (parameter).
    
   
    m = X.shape[0]
    
    # Gradient Descent of loss w.r.t weights.
    dw = (1/m)*np.dot(X.T, (y_sigpred - y))
    
    # Gradient of loss w.r.t bias.
    db = (1/m)*np.sum((y_sigpred - y)) 
    
    return dw, db
```

```python
def decision_boundary(w, b):

    # The Line is y=mx+c
    # So, Equate mx+c = w.X + b
    # Solving we find m and c
    
    m = -w[0]/w[1]
    c = -b/w[1]

    return m,c
```

```python
def train(X, y, bs, iter, lr):
    
    # X --> Input.
    # y --> true/target value.
    # bs --> Batch Size.
    # iter --> Number of iterations.
    # lr --> Learning rate.
        
    #rows,cols
    m, n = X.shape
    
    # Initializing weights and bias to zeros.
    w = np.zeros((n,1))
    b = 0
    
    # Reshaping y.
    y = y.reshape(m,1)
    
   
    
    # Empty list to store losses.
    losses = []
    
    # Training loop.
    for loop_cnt in range(iter):
        
        for i in range((m-1)//bs + 1):
            
            # Defining batches. SGD.
            start_i = i*bs
            end_i = start_i + bs
            xb = X[start_i:end_i]
            yb = y[start_i:end_i]
            
            # Calculating hypothesis/prediction.
            y_sigpred = sigmoid((np.dot(xb, w) + b))
            
            # Getting the gradients of loss w.r.t parameters.
            dw, db = gradients(xb, yb, y_sigpred)
            
            # Updating the parameters.
            w -= lr*dw
            b -= lr*db
        
        # Calculating loss and appending it in the list.
        l = loss(y, sigmoid(np.dot(X, w) + b))
        losses.append(l)
    
        
    # returning weights, bias and losses(List).
    return w, b, losses
```

```python
def predict(X,w,b):
    preds = sigmoid(np.dot(X, w) + b)
    
    pred_class = []
    pred_class = [i for i in preds]
    
    return np.array(pred_class)
```

```python
#one vs All so Iris-virginica as one class and Other two species (Iris-versicolor,Iris-Setosa) as another class
df3 = pd.read_csv("iris.csv")
df3['species'] = df3['species'].replace(['Iris-setosa'],1)
df3['species'] = df3['species'].replace(['Iris-versicolor'],1)
df3['species'] = df3['species'].replace(['Iris-virginica'],0)
sp3=df3.values[:,[0,2]]
op3=df3.values[:,[4]]
op3=op3.flatten()
```

```python
w1, b1, l1 = train(sp1,op1, bs=100, iter=1000, lr=0.01)
m1,c1 = decision_boundary( w1, b1)
```

```python
w3, b3, l3 = train(sp3,op3 ,bs=100, iter=1000, lr=0.01)
m3,c3 = decision_boundary( w3, b3)
```

```python
def three_class_predictor(inp_data,w1,b1,w2,b2,w3,b3):
    pred1= predict(inp_data,w1,b1)[0]
    pred2= predict(inp_data,w2,b2)[0]
    pred3= predict(inp_data,w3,b3)[0]
    pred=[pred1,pred2,pred3]
    pred.sort()
    nearest_zero=pred[0]
   
    if pred1==nearest_zero and pred1<=0.50: 
        return "Iris-setosa"
    elif pred2==nearest_zero and pred2<=0.50:      
        return "Iris-versicolor"
    elif pred3<=0.50:
        return "Iris-virginica"
    
    nearest_zero=pred[1]
    if pred1==nearest_zero:  
        return "Iris-setosa"
    elif pred2==nearest_zero:      
        return "Iris-versicolor"
    else:
        return "Iris-virginica"
    
```

```python
df = pd.read_csv("iris.csv")
output = df.values[:,4]
```

```python
#constructing Confusion Matrix
labels={"Iris-setosa":0,"Iris-versicolor":1,"Iris-virginica":2}
cf=[[0 for i in range(3)] for i in range(3)]
for i in range(150):
    
    true_value=output[i]
    predicted_value=three_class_predictor(sp1[i],w1,b1,w2,b2,w3,b3)
    cf[labels[true_value]][labels[predicted_value]]+=1

```

```python
import seaborn as sns

ax = sns.heatmap(cf, annot=True, cmap='Blues')

ax.set_title('Seaborn Confusion Matrix with labels\n\n');
ax.set_xlabel('\nPredicted Values')
ax.set_ylabel('Actual Values ');

## Ticket labels - List must be in alphabetical order
ax.xaxis.set_ticklabels(['Iris-setosa','Iris-versicolor','Iris-virginica'])
ax.yaxis.set_ticklabels(['Iris-setosa','Iris-versicolor','Iris-virginica'])

## Display the visualization of the Confusion Matrix.
plt.show()
```

```python
#Accuracy Testing
crct=0
for i in range(150):
    if output[i]==three_class_predictor(sp1[i],w1,b1,w2,b2,w3,b3):
        crct+=1

    

print("Accuracy=",crct*100/150)
```
