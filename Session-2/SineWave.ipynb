```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
x = np.arange(-10,10,0.1)
sin_val = np.sin(x)
```

```python
plt.plot(x,sin_val)
plt.title('Sine Wave')
plt.grid(True, which='both')
plt.show()
```

```python
cos_val=np.cos(x)
plt.plot(x,cos_val)
plt.title('Cos Wave')
plt.grid(True, which='both')
plt.show()
```

```python
comb=np.cos(x)+(np.sin(x)**2)+np.tan(x)+(100+3*x+5*x**2)
plt.plot(x,comb)
plt.title('random Func')

plt.show()
```
