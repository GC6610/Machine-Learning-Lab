```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
a,b=map(int,input('Enter the Range').split())
n=int(input("Enter the size of the sequence"))
y = random.sample(range(a,b), n)
x= [i for i in range(n)]
print(y)
```

```python
plt.plot(x,y)
plt.title('X vs Y Function')
plt.show()
```
