```python
#Important features I took here into consideration are Cement, Age
x = df.values[:, [0,7]]  
y = df.values[:, 2]  
m = len(y) #number of rows in the data
```

```python
def Scaling(x):
    mu = np.mean(x, axis = 0)  
    sigma = np.std(x, axis= 0, ddof = 1)  #standard deviation
    x_norm = (x - mu)/sigma
    return x_norm
```

```python
x= Scaling(x)
x
#scaling down the values , i tried this since I was getting overflow error 
```

```python
def compute_cost(x, y, a):
    predictions = x.dot(a)
    dist = (predictions- y) 
    J = 1/(2 * m) * dist.T.dot(dist)
    #cost (1/2*m)summation((pred-y)**2)

    return J
```

```python
def gradient_descent(x, y, a, learning_rate, iterations):
    
    cost_history = []
    #each ai in a ai = ai-(learning_rate*summation_of((partial_derivate wrt ai)))
    #we update for n number of iterations
    for i in range(iterations):
        predictions = x.dot(a)
        dist = predictions- y
        
        sum_delta = (learning_rate / m) * x.transpose().dot(dist)
        a = a - sum_delta
        cost=compute_cost(x, y, a)
        cost_history.append(cost)
        #we travel the cost is decreasing
        if cost>cost_history[-1]:
            break

    return a, cost_history
```

```python
a = np.zeros(2)
iterations = 1000
learning_rate = 0.4
a, cost_history = gradient_descent(x, y, a, learning_rate, iterations)
```
